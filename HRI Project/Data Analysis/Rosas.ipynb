# **Rosas**

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Each Individual User Ratings**

```python
# ROSAS data includes attributes like Perceived Usefulness, Emotional Engagement, Perceived Adaptivity, and Enjoyment for each user.
data_rosas = pd.DataFrame({
    'User': ['Sreeja', 'Pranav', 'Reethika', 'Sharon'],
    'Perceived Usefulness': [4, 3, 4, 3],
    'Emotional Engagement': [5, 4, 3, 4],
    'Perceived Adaptivity': [4, 4, 5, 3],
    'Enjoyment': [3, 3, 4, 3]
})
```

# **ROSAS Visualizations**

```python
# ROSAS data includes attributes like Perceived Usefulness, Emotional Engagement, Perceived Adaptivity, and Enjoyment for each user.
data_rosas = pd.DataFrame({
    'User': ['Sreeja', 'Pranav', 'Reethika', 'Sharon'],
    'Perceived Usefulness': [4, 3, 4, 3],
    'Emotional Engagement': [5, 4, 3, 4],
    'Perceived Adaptivity': [4, 4, 5, 3],
    'Enjoyment': [3, 3, 4, 3]
})

# Set up a general figure size and layout for better alignment and spacing
plt.rcParams['figure.figsize'] = [12, 6]

# Calculate Mean, Mode, and Standard Deviation for each parameter
mean_values_rosas = data_rosas.iloc[:, 1:].mean()
mode_values_rosas = data_rosas.iloc[:, 1:].mode().iloc[0]
std_values_rosas = data_rosas.iloc[:, 1:].std()

# Print Mean, Mode, and Standard Deviation
print("Mean Values (ROSAS):")
print(mean_values_rosas)
print("\nMode Values (ROSAS):")
print(mode_values_rosas)
print("\nStandard Deviation Values (ROSAS):")
print(std_values_rosas)

# 1. Bar Chart for Individual Ratings per User
# This visualization helps in understanding individual user ratings for each parameter
users = data_rosas['User']
attributes = data_rosas.columns[1:]

for user in users:
    user_data = data_rosas[data_rosas['User'] == user].iloc[:, 1:]
    ax = user_data.T.plot(kind='bar', title=f'ROSAS Ratings for {user}', rot=0, legend=False)
    ax.set_xticklabels(attributes)
    plt.xlabel('Attributes')
    plt.ylabel('Rating')
    plt.ylim(0, 5)
    plt.tight_layout(pad=2.0)
    plt.show()

# 2. Grouped Bar Chart to Compare Ratings Across Users
# This visualization helps in comparing ratings for each parameter across users
ax = data_rosas.plot(x='User', kind='bar', title='ROSAS Ratings by User', legend=False)
ax.legend(loc='upper right')
plt.xlabel('User')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

# 3. Heatmap for Correlation Between ROSAS Parameters
# This helps to see the relationships between different dimensions (e.g., how social presence is related to warmth)
plt.figure(figsize=(12, 8))
sns.heatmap(data_rosas.iloc[:, 1:].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Updated ROSAS Dimensions')
plt.tight_layout()
plt.show()

# 4. Box Plot for Distribution of Ratings
# Box plots provide insights into the spread of ratings for each parameter, including median, quartiles, and potential outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=data_rosas.iloc[:, 1:])
plt.title('Distribution of ROSAS Ratings by Parameter')
plt.xlabel('Parameters')
plt.ylabel('Rating')
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Line Chart to Show Trends Across Users
# A line chart to show how each parameter changes across different users
# This visualization can help identify user-specific trends in their perception of the robot
ax = data_rosas.plot(x='User', y=['Perceived Usefulness', 'Emotional Engagement', 'Perceived Adaptivity', 'Enjoyment'], kind='line', marker='o')
plt.title('Trends in ROSAS Ratings Across Users')
plt.xlabel('User')
plt.ylabel('Rating')
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. Average Ratings Bar Chart
# A bar chart for average ratings across all users for each parameter to get an overall impression
ax = mean_values_rosas.plot(kind='bar', title='Average ROSAS Ratings')
plt.xlabel('Parameters')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
