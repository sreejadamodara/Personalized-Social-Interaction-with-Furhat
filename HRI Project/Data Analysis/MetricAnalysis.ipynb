# **Metric Analysis**

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Voice Recognition Accuracy
# Accuracy data for four users (in percentage)
users = ['Sreeja', 'Pranav', 'Sharon', 'Reethika']
correctly_identified = [87, 92, 78, 85]
misidentified = [13, 8, 22, 15]

# Set up the pie chart for voice recognition accuracy
plt.figure(figsize=(10, 6))

# Create pie chart with correctly identified vs misidentified for all users
total_correct = sum(correctly_identified)
total_misidentified = sum(misidentified)
labels = ['Correctly Identified', 'Misidentified']
sizes = [total_correct, total_misidentified]
colors = ['green', 'red']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})

# Add title
plt.title('Overall Voice Recognition Accuracy')

# Display the pie chart
plt.show()

# Create stacked bar chart for voice recognition accuracy per user
plt.figure(figsize=(12, 6))

# Create stacked bar chart with correctly identified vs misidentified for each user
x_pos = np.arange(len(users))
bar_width = 0.5

plt.bar(x_pos, correctly_identified, bar_width, color='green', label='Correctly Identified')
plt.bar(x_pos, misidentified, bar_width, bottom=correctly_identified, color='red', label='Misidentified')

# Add labels and title
plt.xlabel('Users')
plt.ylabel('Accuracy Rate (%)')
plt.title('Voice Recognition Accuracy per User')
plt.xticks(x_pos, users)
plt.ylim(0, 100)
plt.legend()

# Display the stacked bar chart
plt.show()

# 2. Emotion Detection Success Rate
# Emotion detection data
emotions = ['Happy', 'Sad', 'Neutral', 'Angry', 'Surprised', 'Afraid']
success_rate = [80, 70, 85, 75, 60, 65]
missed_rate = [20, 30, 15, 25, 40, 35]

# Set up the plot for emotion detection success rate
plt.figure(figsize=(10, 6))

# Create bar chart for each emotion
y_pos = np.arange(len(emotions))
plt.bar(y_pos, success_rate, color='blue', alpha=0.7, label='Successful Detection')
plt.bar(y_pos, missed_rate, bottom=success_rate, color='orange', alpha=0.7, label='Missed Detection')

# Add labels and title
plt.xlabel('Emotion Type')
plt.ylabel('Detection Rate (%)')
plt.title('Emotion Detection Success Rate for Each Emotion')
plt.xticks(y_pos, emotions)
plt.legend()

# Display the plot
plt.show()

# Create pie chart for emotion detection
plt.figure(figsize=(10, 6))

# Create pie chart for successful vs missed detections across all emotions
total_success = sum(success_rate)
total_missed = sum(missed_rate)
emotion_labels = ['Successful Detection', 'Missed Detection']
emotion_sizes = [total_success, total_missed]
emotion_colors = ['blue', 'orange']

plt.pie(emotion_sizes, labels=emotion_labels, colors=emotion_colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})

# Add title
plt.title('Overall Emotion Detection Feedback')

# Display the pie chart
plt.show()

# 3. Topic Engagement and User Feedback
# Engagement data
topics = ['Health and Wellness', 'Science and Technology', 'Travel and Adventure', 'Movies and Entertainment']
accepted = [50, 60, 40, 55]
declined = [20, 15, 25, 10]

# Set up the plot for topic engagement
plt.figure(figsize=(12, 6))

# Create stacked bar chart for topic engagement
x_pos = np.arange(len(topics))
plt.bar(x_pos, accepted, bar_width, color='green', label='Accepted')
plt.bar(x_pos, declined, bar_width, bottom=accepted, color='red', label='Declined')

# Add labels and title
plt.xlabel('Conversation Topics')
plt.ylabel('Number of Responses')
plt.title('Topic Engagement and User Feedback')
plt.xticks(x_pos, topics)
plt.legend()

# Display the plot
plt.show()

# Create pie chart for topic engagement
plt.figure(figsize=(10, 6))

# Create pie chart for accepted and declined responses across all topics
total_accepted = sum(accepted)
total_declined = sum(declined)
topic_labels = ['Accepted', 'Declined']
topic_sizes = [total_accepted, total_declined]
topic_colors = ['green', 'red']

plt.pie(topic_sizes, labels=topic_labels, colors=topic_colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})

# Add title
plt.title('Overall Topic Engagement Feedback')

# Display the pie chart
plt.show()
```
